```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Seaborn Color Palettes Example

This notebook demonstrates how to use Seaborn's color palettes with the `set_palette` function. Seaborn provides a variety of color palettes that can be used to enhance the visual appeal and interpretability of your plots.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Types of Color Palettes in Seaborn

Seaborn provides several types of color palettes:

1. **Sequential Palettes**: Colors that vary from low to high, suitable for data with a natural ordering
2. **Diverging Palettes**: Colors that diverge from a central point, suitable for data with a meaningful midpoint
3. **Categorical Palettes**: Distinct colors for categorical data
4. **Cubehelix Palettes**: Custom palettes with varying lightness and hue
5. **Custom Palettes**: User-defined color palettes

Let's explore each of these palette types.

## Sequential Palettes

Sequential palettes are suitable for data that has a natural ordering, such as numeric data. Seaborn provides several built-in sequential palettes.

```python
# Create a figure to display sequential palettes
plt.figure(figsize=(12, 8))

# List of sequential palettes
sequential_palettes = ["Blues", "Greens", "Reds", "Purples", "Oranges", "Greys"]

# Display each palette
for i, palette in enumerate(sequential_palettes):
    # Create a color palette with 9 colors
    colors = sns.color_palette(palette, 9)
    
    # Display the palette
    sns.palplot(colors)
    plt.title(palette)
    plt.show()
    
    # Create a simple bar plot with the palette
    plt.figure(figsize=(10, 6))
    with sns.color_palette(palette):
        sns.barplot(x="day", y="total_bill", data=df)
    plt.title(f"Bar Plot with {palette} Palette")
    plt.show()
```

## Diverging Palettes

Diverging palettes are suitable for data that has a meaningful midpoint, such as correlation coefficients or z-scores. Seaborn provides several built-in diverging palettes.

```python
# Create a figure to display diverging palettes
plt.figure(figsize=(12, 8))

# List of diverging palettes
diverging_palettes = ["RdBu", "RdBu_r", "RdYlGn", "RdYlGn_r", "RdYlBu", "RdYlBu_r"]

# Display each palette
for i, palette in enumerate(diverging_palettes):
    # Create a color palette with 9 colors
    colors = sns.color_palette(palette, 9)
    
    # Display the palette
    sns.palplot(colors)
    plt.title(palette)
    plt.show()
    
    # Create a correlation matrix for demonstration
    corr = df.select_dtypes(include='number').corr()
    
    # Create a heatmap with the palette
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, cmap=palette, center=0, annot=True, fmt=".2f")
    plt.title(f"Correlation Heatmap with {palette} Palette")
    plt.show()
```

## Categorical Palettes

Categorical palettes are suitable for categorical data where each category should have a distinct color. Seaborn provides several built-in categorical palettes.

```python
# Create a figure to display categorical palettes
plt.figure(figsize=(12, 8))

# List of categorical palettes
categorical_palettes = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

# Display each palette
for i, palette in enumerate(categorical_palettes):
    # Create a color palette
    colors = sns.color_palette(palette)
    
    # Display the palette
    sns.palplot(colors)
    plt.title(palette)
    plt.show()
    
    # Create a scatter plot with the palette
    plt.figure(figsize=(10, 6))
    with sns.color_palette(palette):
        sns.scatterplot(x="total_bill", y="tip", hue="day", data=df)
    plt.title(f"Scatter Plot with {palette} Palette")
    plt.show()
```

## Cubehelix Palettes

Cubehelix palettes are custom palettes with varying lightness and hue. They are designed to be perceptually uniform in both color and lightness.

```python
# Create a default cubehelix palette
colors = sns.cubehelix_palette()
sns.palplot(colors)
plt.title("Default Cubehelix Palette")
plt.show()

# Create a cubehelix palette with custom parameters
colors = sns.cubehelix_palette(start=0.5, rot=-0.75, dark=0.25, light=0.75, reverse=True, n_colors=9)
sns.palplot(colors)
plt.title("Custom Cubehelix Palette")
plt.show()

# Create a line plot with the custom cubehelix palette
plt.figure(figsize=(10, 6))
with sns.color_palette(colors):
    sns.lineplot(x="size", y="tip", hue="day", data=df)
plt.title("Line Plot with Custom Cubehelix Palette")
plt.show()
```

## Custom Palettes

You can create custom color palettes using RGB tuples, hex color codes, or named colors.

```python
# Create a custom palette using RGB tuples
rgb_palette = [(0.2, 0.4, 0.6), (0.6, 0.4, 0.2), (0.4, 0.6, 0.2)]
sns.palplot(rgb_palette)
plt.title("Custom RGB Palette")
plt.show()

# Create a custom palette using hex color codes
hex_palette = ["#FF5733", "#33FF57", "#3357FF", "#F333FF"]
sns.palplot(sns.color_palette(hex_palette))
plt.title("Custom Hex Palette")
plt.show()

# Create a custom palette using named colors
named_palette = ["red", "green", "blue", "purple", "orange"]
sns.palplot(sns.color_palette(named_palette))
plt.title("Custom Named Palette")
plt.show()

# Create a bar plot with the custom named palette
plt.figure(figsize=(10, 6))
with sns.color_palette(named_palette):
    sns.barplot(x="day", y="total_bill", hue="sex", data=df)
plt.title("Bar Plot with Custom Named Palette")
plt.show()
```

## Setting a Global Palette

You can set a global color palette using the `set_palette` function, which will affect all subsequent plots.

```python
# Set a global palette
sns.set_palette("Set2")

# Create a scatter plot with the global palette
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df)
plt.title("Scatter Plot with Set2 Palette")
plt.show()

# Create a bar plot with the global palette
plt.figure(figsize=(10, 6))
sns.barplot(x="day", y="total_bill", hue="sex", data=df)
plt.title("Bar Plot with Set2 Palette")
plt.show()
```

## Using Palettes Temporarily

You can use the `with` statement to temporarily set a color palette for a specific plot without affecting the global palette.

```python
# Set a global palette
sns.set_palette("Set2")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot with the global palette (Set2)
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df, ax=ax1)
ax1.set_title("Global Palette (Set2)")

# Plot with a temporary palette (Set1)
with sns.color_palette("Set1"):  # Temporarily set the palette to Set1
    sns.scatterplot(x="total_bill", y="tip", hue="day", data=df, ax=ax2)
    ax2.set_title("Temporary Palette (Set1)")

plt.tight_layout()
plt.show()

# Create another plot to show that the global palette is still Set2
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df)
plt.title("Global Palette is Still Set2")
plt.show()
```

## Using Color Maps

In addition to color palettes, you can use color maps for plots like heatmaps and contour plots.

```python
# Create a correlation matrix
corr = df.select_dtypes(include='number').corr()

# List of color maps
cmaps = ["viridis", "plasma", "inferno", "magma", "cividis"]

# Create a heatmap for each color map
for cmap in cmaps:
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, cmap=cmap, annot=True, fmt=".2f")
    plt.title(f"Correlation Heatmap with {cmap} Color Map")
    plt.show()
```

## Combining Styles, Contexts, and Palettes

You can combine different styles, contexts, and palettes to create a wide variety of visual appearances for your plots.

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Different combinations of styles, contexts, and palettes
combinations = [
    ("whitegrid", "paper", "Set1"),
    ("darkgrid", "notebook", "Set2"),
    ("white", "talk", "Set3"),
    ("dark", "poster", "husl")
]

# Create a plot for each combination
for i, (style, context, palette) in enumerate(combinations):
    with sns.axes_style(style), sns.plotting_context(context), sns.color_palette(palette):  # Set style, context, and palette
        # Plot a scatter plot
        row, col = divmod(i, 2)
        sns.scatterplot(data=df, x="total_bill", y="tip", hue="day", ax=axes[row, col])
        axes[row, col].set_title(f"Style: {style}, Context: {context}, Palette: {palette}")

plt.tight_layout()
plt.suptitle("Combining Styles, Contexts, and Palettes", fontsize=16, y=0.98)
plt.show()
```

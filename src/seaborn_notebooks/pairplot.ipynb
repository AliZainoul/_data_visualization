# Pair Plots with Seaborn

Pair plots are a great way to visualize relationships between multiple variables in a dataset. They create a grid of plots where each variable in the dataset is plotted against every other variable. The diagonal of this grid typically shows the distribution of each variable (using histograms or kernel density estimates), while the off-diagonal plots show the relationship between pairs of variables (using scatter plots or other bivariate plots).

Pair plots are particularly useful for:  
- Exploring relationships between multiple variables at once  
- Identifying correlations and patterns in the data  
- Detecting outliers and unusual patterns  
- Getting a quick overview of the dataset  

In Seaborn, pair plots can be created using the `pairplot()` function.

```python
# Load the tips dataset from seaborn
tips = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
tips.head()
```

## Basic `pairplot()` Parameters

The `pairplot()` function in Seaborn has several parameters that allow you to customize the appearance and behavior of the plot:  

- `data`: DataFrame that contains the variables.  
- `hue`: Variable that determines the colors of the plot elements.  
- `vars`: List of variable names to include in the plot.  
- `kind`: Type of plot for the off-diagonal plots ('scatter', 'kde', 'hist', 'reg').  
- `diag_kind`: Type of plot for the diagonal plots ('hist', 'kde').  
- `markers`: Markers to use for the scatter plots when `hue` is used.  
- `height`: Height (in inches) of each facet.  
- `aspect`: Aspect ratio of each facet.  
- `palette`: Palette for mapping the `hue` variable.  
- `corner`: If True, only plots the lower triangle of the grid.  
- `plot_kws`: Dictionary of keyword arguments for the off-diagonal plots.  
- `diag_kws`: Dictionary of keyword arguments for the diagonal plots.  

Let's explore these parameters with examples.

## Basic Pair Plot

Let's start with a basic pair plot that shows the relationships between all numeric variables in the tips dataset.

```python
# Create a basic pair plot
sns.pairplot(tips)
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```

## Pair Plot with Hue Parameter

We can use the `hue` parameter to color the points based on a categorical variable.

```python
# Create a pair plot with hue parameter
sns.pairplot(tips, hue='sex')
plt.suptitle('Pair Plot of Tips Dataset by Sex', y=1.02)
plt.show()
```

## Pair Plot with Selected Variables

We can use the `vars` parameter to select specific variables to include in the plot.

```python
# Create a pair plot with selected variables
sns.pairplot(tips, vars=['total_bill', 'tip', 'size'])
plt.suptitle('Pair Plot of Selected Variables', y=1.02)
plt.show()
```

## Pair Plot with Different Plot Types

We can use the `kind` parameter to change the type of plot used for the off-diagonal plots.

```python
# Create a pair plot with regression plots
sns.pairplot(tips, kind='reg')
plt.suptitle('Pair Plot with Regression Lines', y=1.02)
plt.show()
```

## Pair Plot with Different Diagonal Plot Types

We can use the `diag_kind` parameter to change the type of plot used for the diagonal plots.

```python
# Create a pair plot with KDE plots on the diagonal
sns.pairplot(tips, diag_kind='kde')
plt.suptitle('Pair Plot with KDE Plots on Diagonal', y=1.02)
plt.show()
```

## Pair Plot with Custom Markers

We can use the `markers` parameter to specify different markers for each category when using the `hue` parameter.

```python
# Create a pair plot with custom markers
sns.pairplot(tips, hue='sex', markers=['o', 's'])
plt.suptitle('Pair Plot with Custom Markers', y=1.02)
plt.show()
```

## Pair Plot with Custom Size

We can use the `height` and `aspect` parameters to control the size of the plots.

```python
# Create a pair plot with custom size
sns.pairplot(tips, height=3, aspect=1.2)
plt.suptitle('Pair Plot with Custom Size', y=1.02)
plt.show()
```

## Pair Plot with Custom Palette

We can use the `palette` parameter to specify a custom color palette when using the `hue` parameter.

```python
# Create a pair plot with custom palette
sns.pairplot(tips, hue='time', palette='viridis')
plt.suptitle('Pair Plot with Custom Palette', y=1.02)
plt.show()
```

## Pair Plot with Corner Parameter

We can use the `corner` parameter to only show the lower triangle of the grid, which can be useful for large datasets with many variables.

```python
# Create a pair plot with corner parameter
sns.pairplot(tips, corner=True)
plt.suptitle('Pair Plot with Corner Parameter', y=1.02)
plt.show()
```

## Pair Plot with Custom Plot Parameters

We can use the `plot_kws` and `diag_kws` parameters to pass additional keyword arguments to the off-diagonal and diagonal plots, respectively.

```python
# Create a pair plot with custom plot parameters
sns.pairplot(tips, 
            plot_kws={'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
            diag_kws={'bins': 15, 'alpha': 0.8})
plt.suptitle('Pair Plot with Custom Plot Parameters', y=1.02)
plt.show()
```

## Pair Plot with Iris Dataset

Let's create a pair plot using the famous Iris dataset, which is often used to demonstrate the effectiveness of pair plots for visualizing multivariate data.

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
iris.head()
```

## Advanced Pair Plot Customization

Let's create a more advanced pair plot with multiple customizations.

```python
# Load dataset
iris_raw = load_iris(as_frame=True)
iris = iris_raw.frame
iris['species'] = iris_raw.target_names[iris_raw.target]

# Create an advanced pair plot
g = sns.pairplot(
    iris,
    hue='species',
    palette='viridis',
    diag_kind='kde',
    markers=['o', 's', 'D'],
    height=2.5,
    plot_kws={'alpha': 0.7, 's': 50, 'edgecolor': 'k'},
    diag_kws={'alpha': 0.5, 'linewidth': 2, 'fill': True}  # replaced 'shade' with 'fill'
)

# Customize the plot
g.fig.suptitle('Advanced Pair Plot of Iris Dataset', y=1.02, fontsize=16)

# Adjust tick label size
for ax in g.axes.flatten():
    if ax:
        ax.tick_params(labelsize=12)

# Customize legend
g._legend.set_title('Species')
for text in g._legend.texts:
    text.set_fontsize(12)
g._legend.get_title().set_fontsize(14)

plt.show()
```

## Custom Pair Plot with PairGrid

For even more control over the appearance of the pair plot, we can use Seaborn's `PairGrid` class, which allows us to map different types of plots to different parts of the grid.

```python
# Load iris dataset from seaborn
iris = sns.load_dataset('iris')

g = sns.PairGrid(iris, hue='species', palette='viridis', height=2.5)
g.map_diag(sns.kdeplot, fill=True, alpha=0.5, warn_singular=False)
g.map_upper(sns.scatterplot, s=50, alpha=0.7, edgecolor='k')
g.map_lower(sns.kdeplot, fill=True, alpha=0.5, warn_singular=False)  # removed levels=5

g.add_legend(title='Species', fontsize=12, title_fontsize=14)
g.fig.suptitle('Custom Pair Plot of Iris Dataset', y=1.02, fontsize=16)
plt.show()
```

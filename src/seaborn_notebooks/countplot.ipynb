```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Count Plot Example

This notebook demonstrates how to create count plots using Seaborn's `countplot` function. Count plots are useful for showing the frequency of observations in each categorical bin using bars.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Creating a Basic Count Plot

The `countplot` function in Seaborn creates a count plot with the following key parameters:

- `data`: The DataFrame containing the data
- `x` or `y`: The column name for the categorical variable
- `hue`: Variable that determines the color of bars (optional)

Below, we'll create a count plot showing the number of meals by day of the week, with separate bars for each gender.

```python
# Create a count plot showing the number of meals by day and gender
sns.countplot(data=df, x="day", hue="sex")
plt.title("Number of Meals by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.show()
```

## Vertical vs. Horizontal Count Plots

We can create horizontal count plots by using the `y` parameter instead of `x`. This is particularly useful when dealing with categories with long names.

```python
# Create a horizontal count plot
sns.countplot(data=df, y="day")  # Use y instead of x for horizontal bars
plt.title("Number of Meals by Day (Horizontal)")
plt.xlabel("Count")
plt.ylabel("Day of the Week")
plt.show()
```

## Additional Customization Options

Seaborn's countplot function offers many customization options:

- `order`: Order of the categorical variable levels
- `hue_order`: Order of the hue variable levels
- `palette`: Color palette to use
- `saturation`: Proportion of the original saturation to draw colors at
- `dodge`: Whether to dodge when hue is used (default is True)

Let's create a more customized count plot.

```python
# Create a customized count plot
sns.countplot(
    data=df, 
    x="day", 
    hue="time",  # Color bars by time (lunch/dinner)
    order=["Thur", "Fri", "Sat", "Sun"],  # Custom order of days
    palette="Blues",  # Use the Blues color palette
    saturation=0.8,  # Slightly reduce color saturation
    dodge=True  # Dodge the bars when using hue
)

plt.title("Number of Meals by Day and Time")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.legend(title="Time of Day")
plt.show()
```

## Stacked Bars

We can create stacked bars by setting `dodge=False` when using the `hue` parameter.

```python
# Create a count plot with stacked bars
sns.countplot(
    data=df, 
    x="day", 
    hue="sex",
    dodge=False  # Stack the bars instead of dodging them
)

plt.title("Number of Meals by Day and Gender (Stacked)")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

## Comparing Multiple Categorical Variables

We can use subplots to compare counts across multiple categorical variables.

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Count plot by day
sns.countplot(data=df, x="day", ax=axes[0, 0])
axes[0, 0].set_title("Count by Day")
axes[0, 0].set_xlabel("Day")

# Count plot by time
sns.countplot(data=df, x="time", ax=axes[0, 1])
axes[0, 1].set_title("Count by Time")
axes[0, 1].set_xlabel("Time")

# Count plot by smoker status
sns.countplot(data=df, x="smoker", ax=axes[1, 0])
axes[1, 0].set_title("Count by Smoker Status")
axes[1, 0].set_xlabel("Smoker")

# Count plot by size of party
sns.countplot(data=df, x="size", ax=axes[1, 1])
axes[1, 1].set_title("Count by Party Size")
axes[1, 1].set_xlabel("Size")

plt.tight_layout()
plt.suptitle("Counts of Different Categorical Variables", y=1.02, fontsize=16)
plt.show()
```

## Example with Titanic Dataset

Let's use another dataset to demonstrate count plots. The Titanic dataset contains information about passengers on the Titanic, including whether they survived.

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Display the first few rows
titanic.head()
```

```python
# Create a count plot showing survival by class
sns.countplot(data=titanic, x="class", hue="survived")
plt.title("Survival by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

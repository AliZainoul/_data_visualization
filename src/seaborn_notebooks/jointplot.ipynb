```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Joint Plot Example

This notebook demonstrates how to create joint plots using Seaborn's `jointplot` function. Joint plots show the relationship between two variables along with their individual distributions on the margins.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Creating a Basic Joint Plot

The `jointplot` function in Seaborn creates a multi-panel figure that shows both the joint relationship between two variables and their individual distributions. Key parameters include:

- `data`: The DataFrame containing the data
- `x`: The column name for the x-axis
- `y`: The column name for the y-axis
- `kind`: The kind of plot to draw ('scatter', 'kde', 'hist', 'hex', 'reg', 'resid')

Below, we'll create a joint plot showing the relationship between the total bill and tip amount using a hexbin plot.

```python
# Create a joint plot with hexbin for the central plot
sns.jointplot(data=df, x="total_bill", y="tip", kind="hex")
```

## Different Types of Joint Plots

Seaborn's jointplot function supports several different types of plots for the joint distribution:

- `scatter`: A scatter plot (default)
- `kde`: Kernel density estimate
- `hist`: Histogram
- `hex`: Hexbin plot
- `reg`: Regression plot with marginal histograms
- `resid`: Residual plot

Let's explore some of these different types.

```python
# Create a scatter joint plot (default)
sns.jointplot(data=df, x="total_bill", y="tip", kind="scatter")
plt.suptitle("Scatter Joint Plot", y=1.02)
```

```python
# Create a KDE joint plot
sns.jointplot(data=df, x="total_bill", y="tip", kind="kde")
plt.suptitle("KDE Joint Plot", y=1.02)
```

```python
# Create a regression joint plot
sns.jointplot(data=df, x="total_bill", y="tip", kind="reg")
plt.suptitle("Regression Joint Plot", y=1.02)
```

## Additional Customization Options

Seaborn's jointplot function offers many customization options:

- `height`: Size of the figure (in inches)
- `ratio`: Ratio of joint axis size to marginal axis size
- `space`: Space between the joint and marginal axes
- `color`: Color for all plot elements
- `marginal_kws`: Additional keyword arguments for the marginal plots
- `joint_kws`: Additional keyword arguments for the joint plot

Let's create a more customized joint plot.

```python
# Create a customized joint plot
g = sns.jointplot(
    data=df, 
    x="total_bill", 
    y="tip", 
    kind="scatter",
    height=8,  # Larger figure
    ratio=3,  # Joint plot is 3x the size of the marginal plots
    color="purple",  # Purple color scheme
    joint_kws={"alpha": 0.7, "s": 50},  # Customize scatter points
    marginal_kws={"bins": 20, "fill": True}  # Customize histograms
)

# Add a title to the figure
g.fig.suptitle("Customized Joint Plot of Tips vs Total Bill", y=1.02)

# Add labels to the axes
g.set_axis_labels("Total Bill ($)", "Tip Amount ($)")

# Add a text annotation with statistics
corr = df["total_bill"].corr(df["tip"])
g.ax_joint.text(40, 0.5, f"Correlation: {corr:.2f}", fontsize=12, color="red")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Linear Regression Plot (lmplot) Example

This notebook demonstrates how to create linear regression plots using Seaborn's `lmplot` function. Linear regression plots are useful for visualizing the relationship between two continuous variables and fitting a regression line to show the trend.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Creating a Basic Linear Regression Plot

The `lmplot` function in Seaborn allows us to create scatter plots with a regression line. Key parameters include:

- `data`: The DataFrame containing the data
- `x`: The column name for the x-axis
- `y`: The column name for the y-axis
- `hue`: Variable that determines the color of points and separate regression lines (optional)

Below, we'll create a linear regression plot showing the relationship between the total bill and tip amount, with separate regression lines for each gender.

```python
# Create a linear regression plot showing the relationship between total_bill and tip, with separate lines by sex
sns.lmplot(data=df, x="total_bill", y="tip", hue="sex")
```

## Additional Customization Options

Seaborn's lmplot function offers many customization options:

- `col`: Variable for creating column facets
- `row`: Variable for creating row facets
- `order`: Order of the polynomial regression fit
- `ci`: Confidence interval level for the regression line
- `scatter_kws`: Dictionary of keyword arguments for the scatter plot
- `line_kws`: Dictionary of keyword arguments for the regression line

Let's create a more complex linear regression plot with additional customization.

```python
# Create a more complex linear regression plot with facets and customization
sns.lmplot(
    data=df, 
    x="total_bill", 
    y="tip", 
    hue="sex",  # Separate regression lines by gender
    col="time",  # Create separate columns for lunch and dinner
    height=6,  # Set the height of each facet
    aspect=0.8,  # Set the aspect ratio of each facet
    scatter_kws={"alpha": 0.5, "s": 50},  # Customize scatter points
    line_kws={"linewidth": 2}  # Customize regression line
)

# Add a title to the figure
plt.suptitle("Relationship Between Bill Amount and Tip by Gender and Time", y=1.05)
plt.tight_layout()
```

## Polynomial Regression

We can also fit polynomial regression models by setting the `order` parameter to a value greater than 1.

```python
# Create a polynomial regression plot
sns.lmplot(
    data=df, 
    x="total_bill", 
    y="tip", 
    order=2,  # Fit a quadratic regression line
    line_kws={"color": "red"}
)

plt.title("Quadratic Regression of Tip vs Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")
```

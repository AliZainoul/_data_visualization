```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Box Plots with Seaborn

Box plots (also known as box-and-whisker plots) are a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

Box plots are particularly useful for:  
- Visualizing the distribution of data across different categories  
- Identifying outliers  
- Comparing distributions between groups  
- Showing the central tendency and spread of the data  

In Seaborn, box plots can be created using the `boxplot()` function.

```python
# Load the tips dataset from seaborn
tips = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
tips.head()
```

## Basic `boxplot()` Parameters

The `boxplot()` function in Seaborn has several parameters that allow you to customize the appearance and behavior of the plot:  

- `x`, `y`: Variables that specify positions on the x and y axes.  
- `hue`: Variable that determines the colors of the plot elements.  
- `data`: DataFrame that contains the variables.  
- `order`, `hue_order`: Lists that specify the order of categorical levels.  
- `orient`: Orientation of the plot ('v' for vertical, 'h' for horizontal).  
- `color`: Color for all of the elements.  
- `palette`: Palette for mapping the `hue` variable.  
- `saturation`: Proportion of the original saturation to draw colors.  
- `width`: Width of the boxes.  
- `dodge`: Whether to dodge when `hue` variable is used.  
- `fliersize`: Size of the markers used to indicate outlier observations.  
- `linewidth`: Width of the lines that frame the plot elements.  
- `whis`: Proportion of the IQR past the low and high quartiles to extend the whiskers.  
- `ax`: Axes object to draw the plot onto.  

Let's explore these parameters with examples.

## Basic Box Plot

Let's start with a basic box plot that shows the distribution of the total bill amount.

```python
# Create a basic box plot of the total bill
plt.figure(figsize=(10, 6))
sns.boxplot(x=tips['total_bill'])
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.show()
```

## Box Plot with Categorical Variable

Now, let's create a box plot that shows the distribution of the total bill amount by day of the week.

```python
# Create a box plot of total bill by day
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Distribution of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

## Box Plot with Hue Parameter

We can add another categorical variable using the `hue` parameter to further split the data.

```python
# Create a box plot of total bill by day, split by time (lunch/dinner)
plt.figure(figsize=(12, 6))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.title('Distribution of Total Bill by Day and Time')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.legend(title='Time')
plt.show()
```

## Box Plot with Custom Order

We can specify the order of the categorical variables using the `order` parameter.

```python
# Create a box plot with custom order of days
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Distribution of Total Bill by Day (Custom Order)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

## Box Plot with Custom Palette

We can customize the colors of the boxes using the `palette` parameter.

```python
# Create a box plot with custom palette
plt.figure(figsize=(10, 6))
sns.boxplot(hue='day', y='total_bill', data=tips, palette='Set3')
plt.title('Distribution of Total Bill by Day (Custom Palette)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

## Box Plot with Custom Width

We can adjust the width of the boxes using the `width` parameter.

```python
# Create a box plot with custom width
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, width=0.5)
plt.title('Distribution of Total Bill by Day (Custom Width)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```

## Box Plot with Custom Whiskers

We can adjust the length of the whiskers using the `whis` parameter. By default, the whiskers extend to 1.5 times the interquartile range (IQR). We can change this to show a different range.

```python
# Create a box plot with custom whiskers
plt.figure(figsize=(12, 6))

# Default whiskers (1.5 * IQR)
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Default Whiskers (1.5 * IQR)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

# Custom whiskers (2.0 * IQR)
plt.subplot(1, 2, 2)
sns.boxplot(x='day', y='total_bill', data=tips, whis=2.0)
plt.title('Extended Whiskers (2.0 * IQR)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

plt.tight_layout()
plt.show()
```

## Box Plot with Custom Outlier Size

We can adjust the size of the markers used to indicate outliers using the `fliersize` parameter.

```python
# Create a box plot with custom outlier size
plt.figure(figsize=(12, 6))

# Default outlier size
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Default Outlier Size')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

# Larger outlier size
plt.subplot(1, 2, 2)
sns.boxplot(x='day', y='total_bill', data=tips, fliersize=8)
plt.title('Larger Outlier Size')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

plt.tight_layout()
plt.show()
```

## Horizontal Box Plot

We can create a horizontal box plot by swapping the `x` and `y` parameters or by using the `orient` parameter.

```python
# Create a horizontal box plot
plt.figure(figsize=(10, 6))
sns.boxplot(y='day', x='total_bill', data=tips)
plt.title('Distribution of Total Bill by Day (Horizontal)')
plt.ylabel('Day')
plt.xlabel('Total Bill ($)')
plt.show()
```

## Box Plot with Notches

We can add notches to the box plot using the `notch` parameter. Notches are used to compare groups; if the notches of two boxes do not overlap, this suggests that the medians are significantly different.

```python
# Create a box plot with notches
plt.figure(figsize=(12, 6))

# Default box plot (no notches)
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Default Box Plot (No Notches)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

# Box plot with notches
plt.subplot(1, 2, 2)
sns.boxplot(x='day', y='total_bill', data=tips, notch=True)
plt.title('Box Plot with Notches')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

plt.tight_layout()
plt.show()
```

## Box Plot with Dodge Parameter

When using the `hue` parameter, we can control whether the boxes are dodged (placed side by side) or not using the `dodge` parameter.

```python
# Create a box plot with and without dodge
plt.figure(figsize=(12, 6))

# Default (dodge=True)
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.title('Default (dodge=True)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.legend(title='Time')

# Without dodge (dodge=False)
plt.subplot(1, 2, 2)
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, dodge=False)
plt.title('Without Dodge (dodge=False)')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.legend(title='Time')

plt.tight_layout()
plt.show()
```

## Combining Box Plot with Other Plots

Box plots can be combined with other types of plots to provide more information about the data distribution. For example, we can overlay a strip plot on top of a box plot to show individual data points.

```python
# Create a box plot combined with a strip plot
plt.figure(figsize=(10, 6))

# Create the box plot
sns.boxplot(x='day', y='total_bill', hue='day', data=tips, palette='pastel', legend=False)

# Add a strip plot on top
sns.stripplot(x='day', y='total_bill', data=tips, color='black', alpha=0.5, jitter=True, dodge=True)

plt.title('Box Plot with Strip Plot Overlay')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.tight_layout()
plt.show()
```

## Combining Box Plot with Swarm Plot

We can also combine a box plot with a swarm plot, which is similar to a strip plot but avoids overlapping points.

```python
plt.figure(figsize=(10, 6))

# Create the box plot with explicit hue to avoid FutureWarning
sns.boxplot(x='day', y='total_bill', hue='day', data=tips, palette='pastel', legend=False)

# Add a swarm plot on top
sns.swarmplot(x='day', y='total_bill', data=tips, color='black', alpha=0.5)

plt.title('Box Plot with Swarm Plot Overlay')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.tight_layout()
plt.show()
```

## Comparing Box Plot with Violin Plot

Let's compare a box plot with a violin plot, which shows the distribution of the data using kernel density estimation.

```python
# Compare box plot with violin plot
plt.figure(figsize=(12, 6))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

# Violin plot
plt.subplot(1, 2, 2)
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')

plt.tight_layout()
plt.show()
```

## Advanced Box Plot Customization

Let's create a more advanced box plot with multiple customizations.

```python
plt.figure(figsize=(12, 8))

# Create a custom color palette with exactly 2 colors for the 'time' hue
custom_palette = sns.color_palette("husl", 2)

# Create the box plot with multiple customizations
sns.boxplot(
    x='day', y='total_bill', hue='time', data=tips,
    palette=custom_palette, width=0.7, fliersize=5,
    linewidth=1.5, notch=True, order=['Thur', 'Fri', 'Sat', 'Sun']
)

# Add a strip plot for individual data points
sns.stripplot(
    x='day', y='total_bill', hue='time', data=tips,
    dodge=True, alpha=0.5, palette='dark:black', size=4,
    order=['Thur', 'Fri', 'Sat', 'Sun']
)

# Customize the plot
plt.title('Distribution of Total Bill by Day and Time', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Total Bill ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adjust the legend
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[:2], labels[:2], title='Time', fontsize=12, title_fontsize=14)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Using `catplot()` for Box Plots

Seaborn's `catplot()` function can be used to create box plots in a figure-level interface, which is useful for creating faceted plots.

```python
# Create a faceted box plot using catplot
g = sns.catplot(x='day', y='total_bill', hue='sex', col='time',
                data=tips, kind='box', height=6, aspect=0.7,
                palette='viridis', legend_out=False)

# Customize the plot
g.set_axis_labels("Day of the Week", "Total Bill ($)")
g.set_titles("Time: {col_name}")
g.fig.suptitle('Distribution of Total Bill by Day, Sex, and Time', y=1.02, fontsize=16)
g.tight_layout()

plt.show()
```

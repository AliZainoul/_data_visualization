```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Seaborn Styles Example

This notebook demonstrates how to use Seaborn's predefined styles with the `set_style` function. Seaborn provides several built-in themes that control the appearance of plots, making it easy to create visually appealing and consistent visualizations.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Available Styles in Seaborn

Seaborn provides five predefined styles that can be set using the `set_style` function:

1. `darkgrid`: Dark background with grid lines (default)
2. `whitegrid`: White background with grid lines
3. `dark`: Dark background without grid lines
4. `white`: White background without grid lines
5. `ticks`: White background with axis ticks

Let's explore each of these styles by creating the same plot with different styles.

### Darkgrid Style (Default)

The darkgrid style is the default style in Seaborn. It features a dark background with grid lines, which is useful for data-heavy plots.

```python
# Set the style to darkgrid (default)
sns.set_style("darkgrid")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Darkgrid Style (Default)")
plt.show()
```

### Whitegrid Style

The whitegrid style features a white background with grid lines. This style is often preferred for presentations and publications.

```python
# Set the style to whitegrid
sns.set_style("whitegrid")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Whitegrid Style")
plt.show()
```

### Dark Style

The dark style features a dark background without grid lines. This style is useful when you want to emphasize the data points themselves.

```python
# Set the style to dark
sns.set_style("dark")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Dark Style")
plt.show()
```

### White Style

The white style features a white background without grid lines. This style is clean and minimalist, focusing attention on the data.

```python
# Set the style to white
sns.set_style("white")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("White Style")
plt.show()
```

### Ticks Style

The ticks style features a white background with axis ticks. This style is similar to the default style in Matplotlib.

```python
# Set the style to ticks
sns.set_style("ticks")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Ticks Style")
plt.show()
```

## Comparing All Styles

Let's create a figure with subplots to compare all five styles side by side.

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
axes = axes.flatten()

# List of Seaborn styles
styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

for i, style in enumerate(styles):
    sns.set_style(style)  # Set seaborn style (not plt.style.context)
    sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=axes[i])
    axes[i].set_title(f"{style.capitalize()} Style")
    axes[i].set_xlabel("Total Bill ($)")
    axes[i].set_ylabel("Tip Amount ($)")

# Hide the last subplot (empty)
axes[5].set_visible(False)

plt.tight_layout()
plt.suptitle("Comparison of Seaborn Styles", fontsize=16, y=0.98)
plt.show()
```

## Customizing Styles

In addition to the predefined styles, you can customize specific style elements using the `rc` parameter in `set_style` or by using the `set` function.

```python
# Set the base style to whitegrid
sns.set_style("whitegrid", {
    "axes.facecolor": "#f0f0f0",  # Light gray background
    "grid.color": "white",  # White grid lines
    "grid.linestyle": "--",  # Dashed grid lines
    "axes.edgecolor": "gray",  # Gray edge color
    "axes.linewidth": 1.2  # Thicker axes lines
})

# Create a scatter plot with the customized style
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Customized Whitegrid Style")
plt.show()
```

## Removing Spines with despine()

Seaborn provides the `despine()` function to remove the top and right spines of a plot, which is a common practice in data visualization to reduce chart junk.

```python
# Set the style to whitegrid
sns.set_style("whitegrid")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot without despine
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=ax1)
ax1.set_title("Without despine()")

# Plot with despine
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=ax2)
sns.despine(ax=ax2)  # Remove top and right spines
ax2.set_title("With despine()")

plt.tight_layout()
plt.show()
```

### Customizing despine()

The `despine()` function offers several parameters to customize which spines to remove and how they should be removed.

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()

# Basic despine (removes top and right)
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=axes[0])
sns.despine(ax=axes[0])
axes[0].set_title("Basic despine()")

# Remove all spines
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=axes[1])
sns.despine(ax=axes[1], left=True, bottom=True)  # Remove all spines
axes[1].set_title("despine(left=True, bottom=True)")

# Offset spines from the axes
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=axes[2])
sns.despine(ax=axes[2], offset=10)  # Offset spines by 10 points
axes[2].set_title("despine(offset=10)")

# Trim spines to the data range
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=axes[3])
sns.despine(ax=axes[3], trim=True)  # Trim spines to the data range
axes[3].set_title("despine(trim=True)")

plt.tight_layout()
plt.suptitle("Customizing despine()", fontsize=16, y=0.98)
plt.show()
```

## Using Styles Temporarily

You can use the `with` statement to temporarily set a style for a specific plot without affecting the global style.

```python
# Set the global style to whitegrid
sns.set_style("whitegrid")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot with the global style (whitegrid)
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=ax1)
ax1.set_title("Global Style (whitegrid)")

# Plot with a temporary style (dark)
with sns.axes_style("dark"):  # Temporarily set the style to dark
    sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=ax2)
    ax2.set_title("Temporary Style (dark)")

plt.tight_layout()
plt.show()

# Create another plot to show that the global style is still whitegrid
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Global Style is Still Whitegrid")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Scatterplot Example

This notebook demonstrates how to create a scatterplot using Seaborn. Scatterplots are useful for visualizing the relationship between two continuous variables.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Creating a Basic Scatterplot

The `scatterplot` function in Seaborn allows us to create scatter plots with the following key parameters:

- `data`: The DataFrame containing the data
- `x`: The column name for the x-axis
- `y`: The column name for the y-axis
- `hue`: Variable that determines the color of points (optional)

Below, we'll create a scatterplot showing the relationship between the total bill and tip amount, with points colored by gender.

```python
# Create a scatterplot showing the relationship between total_bill and tip, colored by sex
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Pourboire en fonction de l'addition")
plt.show()
```

## Additional Customization Options

Seaborn's scatterplot function offers many customization options:

- `size`: Variable that determines the size of points
- `style`: Variable that determines the style of points
- `palette`: Color palette to use
- `alpha`: Opacity of the points

Let's create a more complex scatterplot with additional variables.

```python
# Create a more complex scatterplot with additional variables
sns.scatterplot(
    data=df, 
    x="total_bill", 
    y="tip", 
    hue="day",  # Color points by day of the week
    size="size",  # Size points by party size
    style="smoker",  # Different marker styles for smokers/non-smokers
    alpha=0.7  # Slightly transparent points
)

plt.title("Relationship Between Bill Amount and Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")
plt.show()
```

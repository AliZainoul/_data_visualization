```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Strip Plot Example

This notebook demonstrates how to create strip plots using Seaborn's `stripplot` function. Strip plots show the distribution of data points along a categorical axis, with each point representing an individual observation.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Creating a Basic Strip Plot

The `stripplot` function in Seaborn creates a strip plot with the following key parameters:

- `data`: The DataFrame containing the data
- `x`: The column name for the categorical variable (x-axis)
- `y`: The column name for the continuous variable (y-axis)
- `hue`: Variable that determines the color of points (optional)
- `jitter`: Whether to jitter the points along the categorical axis (default is True)

Below, we'll create a strip plot showing the distribution of total bill amounts by day of the week.

```python
# Create a basic strip plot
sns.stripplot(data=df, x="day", y="total_bill")
plt.title("Distribution of Bill Amounts by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.show()
```

## Understanding Jitter

By default, Seaborn's stripplot adds a small amount of random noise to the categorical positions (jitter) to avoid overlapping points. We can control this behavior using the `jitter` parameter:

- `jitter=True`: Add jitter (default)
- `jitter=False`: No jitter
- `jitter=0.25`: Control the amount of jitter (a float between 0 and 1)

Let's compare different jitter settings.

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Strip plot with no jitter
sns.stripplot(data=df, x="day", y="total_bill", jitter=False, ax=axes[0])
axes[0].set_title("No Jitter")

# Strip plot with default jitter
sns.stripplot(data=df, x="day", y="total_bill", jitter=True, ax=axes[1])
axes[1].set_title("Default Jitter")

# Strip plot with maximum jitter
sns.stripplot(data=df, x="day", y="total_bill", jitter=1, ax=axes[2])
axes[2].set_title("Maximum Jitter")

plt.tight_layout()
plt.suptitle("Effect of Jitter on Strip Plots", y=1.05, fontsize=16)
plt.show()
```

## Combining Strip Plots with Violin Plots

A common technique is to combine strip plots with violin plots to show both the individual data points and the distribution. This provides a comprehensive view of the data.

```python
# Create a violin plot with no inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner=None)

# Add a strip plot on top
sns.stripplot(data=df, x="day", y="total_bill", color="k", alpha=0.3)

plt.title("Distribution by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.show()
```

## Additional Customization Options

Seaborn's stripplot function offers many customization options:

- `hue`: Variable that determines the color of points
- `dodge`: Whether to dodge when hue is used (separate points by hue)
- `order`: Order of the categorical variable levels
- `hue_order`: Order of the hue variable levels
- `palette`: Color palette to use
- `size`: Size of the markers
- `marker`: Marker style
- `alpha`: Transparency of the markers

Let's create a more customized strip plot.

```python
# Create a customized strip plot
sns.stripplot(
    data=df, 
    x="day", 
    y="total_bill", 
    hue="sex",  # Color points by gender
    dodge=True,  # Dodge the points by hue
    palette="Set2",  # Use the Set2 color palette
    size=8,  # Larger markers
    marker="o",  # Circle markers
    alpha=0.7,  # Slightly transparent
    jitter=0.3,  # Control the amount of jitter
    order=["Thur", "Fri", "Sat", "Sun"]  # Custom order of days
)

plt.title("Distribution of Bill Amounts by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.legend(title="Gender")
plt.show()
```

## Using Different Markers

We can use different marker styles to represent different categories in the data.

```python
sns.scatterplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex",          # color by gender
    style="sex",        # different markers by gender
    alpha=0.7,
    s=100               # size of markers
)

plt.title("Distribution of Bill Amounts by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.legend(title="Gender")
plt.show()
```

## Horizontal Strip Plots

We can create horizontal strip plots by swapping the `x` and `y` parameters.

```python
# Create a horizontal strip plot
sns.stripplot(
    data=df, 
    y="day",  # Categorical variable on y-axis for horizontal strips
    x="total_bill",  # Continuous variable on x-axis
    hue="time",  # Color by time of day
    palette="pastel"
)

plt.title("Distribution of Bill Amounts by Day and Time (Horizontal)")
plt.xlabel("Total Bill Amount ($)")
plt.ylabel("Day of the Week")
plt.legend(title="Time")
plt.tight_layout()
plt.show()
```

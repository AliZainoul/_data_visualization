```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Violin Plot Example

This notebook demonstrates how to create violin plots using Seaborn's `violinplot` function. Violin plots are a combination of box plots and kernel density plots, showing the distribution of data across different categories.

```python
# Load the tips dataset from seaborn
df = sns.load_dataset("tips")

# Display the first few rows of the dataset to understand its structure
df.head()
```

## Creating a Basic Violin Plot

The `violinplot` function in Seaborn creates a violin plot with the following key parameters:

- `data`: The DataFrame containing the data
- `x`: The column name for the categorical variable (x-axis)
- `y`: The column name for the continuous variable (y-axis)
- `hue`: Variable that determines the color of violins (optional)
- `inner`: Inner visual representation ('box', 'quartile', 'point', 'stick', None)

Below, we'll create a violin plot showing the distribution of total bill amounts by day of the week.

```python
# Create a basic violin plot
sns.violinplot(data=df, x="day", y="total_bill")
plt.title("Distribution of Bill Amounts by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.show()
```

## Inner Representations

Seaborn's violinplot function offers several options for the inner representation using the `inner` parameter:

- `inner='box'`: Show a miniature box plot inside each violin
- `inner='quartile'`: Show the quartiles of the distribution
- `inner='point'`: Show a central point (usually the median)
- `inner='stick'`: Show a stick representing the individual observations
- `inner=None`: No inner representation

Let's compare different inner representations.

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Violin plot with box inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner="box", ax=axes[0, 0])
axes[0, 0].set_title("inner='box'")

# Violin plot with quartile inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner="quartile", ax=axes[0, 1])
axes[0, 1].set_title("inner='quartile'")

# Violin plot with point inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner="point", ax=axes[0, 2])
axes[0, 2].set_title("inner='point'")

# Violin plot with stick inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner="stick", ax=axes[1, 0])
axes[1, 0].set_title("inner='stick'")

# Violin plot with no inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner=None, ax=axes[1, 1])
axes[1, 1].set_title("inner=None")

# Hide the last subplot (empty)
axes[1, 2].set_visible(False)

plt.tight_layout()
plt.suptitle("Different Inner Representations in Violin Plots", y=1.02, fontsize=16)
plt.show()
```

## Combining Violin Plots with Strip Plots

A common technique is to combine violin plots with strip plots to show both the distribution and the individual data points. This is especially useful when the inner representation is set to None.

```python
# Create a violin plot with no inner representation
sns.violinplot(data=df, x="day", y="total_bill", inner=None)

# Add a strip plot on top
sns.stripplot(data=df, x="day", y="total_bill", color="k", alpha=0.3)

plt.title("Distribution by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.show()
```

## Additional Customization Options

Seaborn's violinplot function offers many customization options:

- `hue`: Variable that determines the color of violins
- `split`: Whether to split the violins when hue is used
- `scale`: Method to scale the width of each violin ('area', 'count', 'width')
- `order`: Order of the categorical variable levels
- `hue_order`: Order of the hue variable levels
- `palette`: Color palette to use
- `bw`: Controls the smoothness of the violin

Let's create a more customized violin plot.

```python
# Create a customized violin plot
sns.violinplot(
    data=df, 
    x="day", 
    y="total_bill", 
    hue="sex",  # Color violins by gender
    split=True,  # Split the violins by hue
    inner="quartile",  # Show quartiles inside the violins
    palette="Set2",  # Use the Set2 color palette
    density_norm="width",
    order=["Thur", "Fri", "Sat", "Sun"]  # Custom order of days
)

plt.title("Distribution of Bill Amounts by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill Amount ($)")
plt.legend(title="Gender")
plt.show()
```

## Horizontal Violin Plots

We can create horizontal violin plots by swapping the `x` and `y` parameters.

```python
# Create a horizontal violin plot
sns.violinplot(
    data=df, 
    y="day",
    x="total_bill",
    hue="day",       # assign 'day' also to hue to apply the palette
    palette="pastel",
    legend=False     # hide the legend
)


plt.title("Distribution of Bill Amounts by Day (Horizontal)")
plt.xlabel("Total Bill Amount ($)")
plt.ylabel("Day of the Week")
plt.tight_layout()
plt.show()
```

# Exercise 2: Histogram with density curve


```python
import matplotlib.pyplot as plt  # https://matplotlib.org/stable/api/pyplot_api.html
import numpy as np               # https://numpy.org/doc/
import seaborn as sns           # https://seaborn.pydata.org/

# -----------------------------------------------------------------------------
# Set the random seed for reproducibility
# -----------------------------------------------------------------------------
np.random.seed(1)
# numpy.random.seed(seed): Fixes the seed for consistent random values across runs
# Docs: https://numpy.org/doc/stable/reference/random/generated/numpy.random.seed.html

# -----------------------------------------------------------------------------
# Generate 1000 samples from a normal distribution (mean=0, std=1)
# -----------------------------------------------------------------------------
data = np.random.normal(size=1000)
# numpy.random.normal(): Draw samples from a normal distribution
# Docs: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html

# -----------------------------------------------------------------------------
# Create a new figure for plotting
# -----------------------------------------------------------------------------
plt.figure(figsize=(8, 6))
# matplotlib.pyplot.figure(): Initializes a new figure
# figsize sets width and height in inches
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

# -----------------------------------------------------------------------------
# Draw a histogram with a KDE (Kernel Density Estimate) overlay
# -----------------------------------------------------------------------------
sns.histplot(
    data=data,        # 1D array of data points
    kde=True,         # Enable the kernel density estimate line
    bins=30,          # Number of histogram bins
    color='skyblue'   # Color of the bars
)
# seaborn.histplot(): Draws a histogram with optional KDE
# Docs: https://seaborn.pydata.org/generated/seaborn.histplot.html

# -----------------------------------------------------------------------------
# Set plot title
# -----------------------------------------------------------------------------
plt.title("Histogram with KDE")
# matplotlib.pyplot.title(): Set the title of the current axes
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html

# -----------------------------------------------------------------------------
# Label the x-axis
# -----------------------------------------------------------------------------
plt.xlabel("Value")
# matplotlib.pyplot.xlabel(): Label for the x-axis
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html

# -----------------------------------------------------------------------------
# Label the y-axis
# -----------------------------------------------------------------------------
plt.ylabel("Density")
# matplotlib.pyplot.ylabel(): Label for the y-axis
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html

# -----------------------------------------------------------------------------
# Display the final plot
# -----------------------------------------------------------------------------
plt.show()
# matplotlib.pyplot.show(): Render the plot
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
```

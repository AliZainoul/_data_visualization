# Exercise 5: 3D surface plot


```python
import numpy as np  # https://numpy.org/doc/
import matplotlib.pyplot as plt  # https://matplotlib.org/stable/api/pyplot_api.html
from mpl_toolkits.mplot3d import Axes3D  # Enables 3D plotting

# -----------------------------------------------------------------------------
# Create a new figure with custom size
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(10, 7))
# matplotlib.pyplot.figure(): Creates a new figure
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

# -----------------------------------------------------------------------------
# Add a 3D subplot to the figure
# -----------------------------------------------------------------------------
ax = fig.add_subplot(111, projection='3d')
# figure.add_subplot(): Adds a subplot to the figure
# projection='3d' enables 3D plotting via mplot3d
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.add_subplot.html

# -----------------------------------------------------------------------------
# Create meshgrid data
# -----------------------------------------------------------------------------
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
# numpy.linspace(): Generates evenly spaced numbers over a range
# Docs: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

X, Y = np.meshgrid(x, y)
# numpy.meshgrid(): Returns coordinate matrices from coordinate vectors
# Docs: https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html

# -----------------------------------------------------------------------------
# Compute Z values using a radial sine function
# -----------------------------------------------------------------------------
Z = np.sin(np.sqrt(X**2 + Y**2))
# numpy.sqrt(), numpy.sin(): Apply square root and sine element-wise
# Docs: https://numpy.org/doc/stable/reference/generated/numpy.sqrt.html
# Docs: https://numpy.org/doc/stable/reference/generated/numpy.sin.html

# -----------------------------------------------------------------------------
# Plot a 3D surface using X, Y, Z
# -----------------------------------------------------------------------------
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
# ax.plot_surface(): Plots a 3D surface
# cmap='viridis' defines the color map
# Docs: https://matplotlib.org/stable/api/_as_gen/mpl_toolkits.mplot3d.axes3d.Axes3D.plot_surface.html

# -----------------------------------------------------------------------------
# Add a colorbar to map colors to Z values
# -----------------------------------------------------------------------------
fig.colorbar(surf)
# figure.colorbar(): Adds a colorbar to the figure
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.colorbar.html

# -----------------------------------------------------------------------------
# Add a title to the figure
# -----------------------------------------------------------------------------
plt.title("3D Surface Plot of sin(sqrt(x^2+y^2))")
# matplotlib.pyplot.title(): Sets a title for the current axes
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html

# -----------------------------------------------------------------------------
# Display the final 3D plot
# -----------------------------------------------------------------------------
plt.show()
# matplotlib.pyplot.show(): Displays the plot in a window
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
```

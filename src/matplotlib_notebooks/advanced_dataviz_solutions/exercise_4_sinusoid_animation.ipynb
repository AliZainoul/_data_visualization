# Exercise 4: Sinusoid animation


```python
"""
Animated Sine Wave with Matplotlib's FuncAnimation

This script generates a smooth, real-time animation of a sine wave
with a dynamic phase shift using `matplotlib.animation.FuncAnimation`.

It is designed for use in Jupyter Notebooks and produces an HTML5-based animation
that can be played inline.

---

Key Features:
- Dynamic update of sine wave phase
- Uses `FuncAnimation` for efficient animation handling
- Compatible with HTML output in Jupyter environments

References:
- FuncAnimation: https://matplotlib.org/stable/api/animation_api.html
- HTML rendering: https://matplotlib.org/stable/gallery/animation/animated_figure.html
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# -----------------------------------------------------------------------------
# Step 1: Create the figure and axis for plotting
# -----------------------------------------------------------------------------
fig, ax = plt.subplots()
# plt.subplots(): Creates a figure and a single subplot.
# Doc: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

x = np.linspace(0, 2 * np.pi, 100)
# np.linspace(): Generates 100 linearly spaced values between 0 and 2π.
# Doc: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

line, = ax.plot(x, np.sin(x))
# ax.plot(): Draws the initial sine wave.
# The comma unpacks the Line2D object (ax.plot returns a tuple).

ax.set_ylim(-1.5, 1.5)
# ax.set_ylim(): Sets the fixed y-axis range.
# Doc: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylim.html

# -----------------------------------------------------------------------------
# Step 2: Define the update function for the animation
# -----------------------------------------------------------------------------
def update(frame):
    """
    Update function for FuncAnimation.
    
    Parameters:
    - frame (int): The current frame number (automatically provided)
    
    Returns:
    - tuple: The updated matplotlib Line2D object
    """
    line.set_ydata(np.sin(x + frame / 10))  # Shift the sine wave horizontally
    return line,

# -----------------------------------------------------------------------------
# Step 3: Create the animation object
# -----------------------------------------------------------------------------
ani = FuncAnimation(
    fig=fig,             # The matplotlib figure to animate
    func=update,         # The update function for each frame
    frames=100,          # Total number of frames
    interval=50          # Delay between frames in milliseconds
)
# Doc: https://matplotlib.org/stable/api/_as_gen/matplotlib.animation.FuncAnimation.html

# -----------------------------------------------------------------------------
# Step 4: Display the animation as HTML5 video
# -----------------------------------------------------------------------------
HTML(ani.to_jshtml())
# to_jshtml(): Renders the animation in a Jupyter Notebook using JavaScript.
# Doc: https://matplotlib.org/stable/users/prev_whats_new/whats_new_3.1.0.html#improved-javascript-rendering

# Optional alternative if you want to view it statically (not recommended for animation):
# plt.show()


# Steps:
# Step 1: run the script in a Jupyter Notebook.
# Step 2: The animation will be displayed as an HTML5 video.
# Step 3: Click on the play button to start the animation emoji play button ▶️.
```

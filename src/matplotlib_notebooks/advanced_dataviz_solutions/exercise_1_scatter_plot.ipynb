# Exercise 1: Scatter plot


```python
import matplotlib.pyplot as plt  # https://matplotlib.org/stable/api/pyplot_api.html
import numpy as np  # https://numpy.org/doc/

# -----------------------------------------------------------------------------
# Set the random seed for reproducibility
# -----------------------------------------------------------------------------
np.random.seed(0)
# numpy.random.seed(seed): Fixes the seed for random number generation.
# Ensures that the same random numbers are generated every time the script runs.
# Docs: https://numpy.org/doc/stable/reference/random/generated/numpy.random.seed.html

# -----------------------------------------------------------------------------
# Generate random data for the scatter plot
# -----------------------------------------------------------------------------
x = np.random.rand(100)
# numpy.random.rand(n): Returns n samples from a uniform distribution over [0, 1).
# Docs: https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html

y = np.random.rand(100)
colors = np.random.rand(100)  # One scalar value per point for coloring
sizes = 1000 * np.random.rand(100)  # Variable size for each point (scaled up)

# -----------------------------------------------------------------------------
# Create a new figure with a custom size
# -----------------------------------------------------------------------------
plt.figure(figsize=(8, 6))
# matplotlib.pyplot.figure(): Creates a new figure.
# figsize=(width, height) in inches.
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

# -----------------------------------------------------------------------------
# Create a scatter plot with transparency, color mapping, and point size
# -----------------------------------------------------------------------------
sc = plt.scatter(
    x,                # X-coordinates of points
    y,                # Y-coordinates of points
    c=colors,         # Color values (can be a list or array)
    s=sizes,          # Size of each point
    alpha=0.6,        # Transparency (0 = fully transparent, 1 = opaque)
    cmap='viridis'    # Color map used to map scalar data to colors
)
# matplotlib.pyplot.scatter(): Creates a scatter plot.
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
# Color maps: https://matplotlib.org/stable/tutorials/colors/colormaps.html

# -----------------------------------------------------------------------------
# Add a colorbar to show color scale
# -----------------------------------------------------------------------------
plt.colorbar(sc)
# matplotlib.pyplot.colorbar(mappable): Adds a colorbar to the plot.
# 'sc' is the mappable object returned by scatter().
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.colorbar.html

# -----------------------------------------------------------------------------
# Set the title of the plot
# -----------------------------------------------------------------------------
plt.title("Random Scatter Plot")
# matplotlib.pyplot.title(label): Sets the title of the current axes.
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html

# -----------------------------------------------------------------------------
# Show the plot window
# -----------------------------------------------------------------------------
plt.show()
# matplotlib.pyplot.show(): Displays all open figures.
# Required in scripts to render the figure.
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
```

# Exercise 3: Multiple plots with subplot


```python
import matplotlib.pyplot as plt  # https://matplotlib.org/stable/api/pyplot_api.html
import numpy as np               # https://numpy.org/doc/

# -----------------------------------------------------------------------------
# Generate 400 points between -2 and 2
# -----------------------------------------------------------------------------
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
# numpy.linspace(start, stop, num): Creates evenly spaced values over a range
# Docs: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

# -----------------------------------------------------------------------------
# Create a 2x2 grid of subplots with a specified figure size
# -----------------------------------------------------------------------------
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# matplotlib.pyplot.subplots(): Returns a figure and a grid of Axes
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

# -----------------------------------------------------------------------------
# Top-left subplot: sin(x)
# -----------------------------------------------------------------------------
axs[0, 0].plot(x, np.sin(x))
# ax.plot(): Plots y versus x as lines and/or markers
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html
axs[0, 0].set_title("sin(x)")
# ax.set_title(): Sets the title of the subplot
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html

# -----------------------------------------------------------------------------
# Top-right subplot: cos(x)
# -----------------------------------------------------------------------------
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("cos(x)")

# -----------------------------------------------------------------------------
# Bottom-left subplot: tan(x) with limited y-axis to avoid vertical asymptotes
# -----------------------------------------------------------------------------
axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_ylim(-5, 5)
# ax.set_ylim(): Sets the y-axis limits to improve readability
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylim.html
axs[1, 0].set_title("tan(x)")

# -----------------------------------------------------------------------------
# Bottom-right subplot: exp(x/3)
# -----------------------------------------------------------------------------
axs[1, 1].plot(x, np.exp(x / 3))
axs[1, 1].set_title("exp(x/3)")
# numpy.exp(): Calculates the exponential of all elements in the input array
# Docs: https://numpy.org/doc/stable/reference/generated/numpy.exp.html

# -----------------------------------------------------------------------------
# Automatically adjust subplot layout to prevent overlapping elements
# -----------------------------------------------------------------------------
plt.tight_layout()
# matplotlib.pyplot.tight_layout(): Adjusts subplots to fit in figure area
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html

# -----------------------------------------------------------------------------
# Render the entire figure
# -----------------------------------------------------------------------------
plt.show()
# matplotlib.pyplot.show(): Display the figure
# Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
```

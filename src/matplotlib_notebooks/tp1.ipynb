# TP 1 - Marks over weeks

```python
# 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2. Simulating data via Python lists
marks = [12, 14, 15, 13, 16]
weeks = ["W1", "W2", "W3", "W4", "W5"]
```

```python
# 3. Plot a line chart with 'weeks' on the x-axis and 'marks' on the y-axis
plt.plot(weeks, marks)
```

```python
# 4. Show the plot
plt.plot(weeks, marks)
# Set the x-axis label
plt.xlabel("Weeks")
# Set the y-axis label
plt.ylabel("Marks")
# Set the title of the plot
plt.title("Marks Over Weeks")
plt.show()
```
